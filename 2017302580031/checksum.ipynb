```python
import numpy as np
from matplotlib import pyplot as plt

def CheckSum(bit_list):
    sum = 0
    for bit in bit_list:
        sum += int(bit,2)
        if sum > 0xffff:
            sum = (sum & 0xffff) + (sum >> 16)
    sum = 0xffff - sum
    return sum

def showplt(bit_list,sum):
    plt.title("matplot")
    plt.xlabel("bit")
    plt.ylabel("value")
    for index in range(len(bit_list)):
        x = []
        y = []
        for i in range(16):
            x.append(i+1)
            y.append(int(bit_list[index][i]))
        plt.plot(x, y,label="num{0}".format(index+1))
    x = []
    y = []
    for i in range(16):
        x.append(i + 1)
        y.append(int(sum[i]))
    plt.plot(x, y, label="sum")
    plt.legend(loc='upper left')
    plt.show()

if __name__ == '__main__':
    list = ["0110011001100000","0101010101010101","1000111100001100"]
    result = CheckSum(list)
    for i in range(len(list)):
        print("第{0}个检验数字为：{1}".format(i+1,list[i]))
    print("真实答案为1011010100111101")
    print("检验和为  "+np.binary_repr(result,width=16))
    showplt(list,str(np.binary_repr(CheckSum(bit_list),16)))
```
